## 5. Model Building

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise  import linear_kernel
```

```python
tfid=TfidfVectorizer()
```

```python
vectors = tfid.fit_transform(movies_data['Genre'])
```

```python
vectors[0].toarray()
```

```python
sim_distance = linear_kernel(vectors[0],vectors)
sim_distance
```

```python
sim_score_df=pd.DataFrame(sim_distance).T
sim_score_df.columns=['Scores']
```

```python
sim_score_df
```

```python
index =pd.Series(data=movies_data.index,
    index=movies_data["Title"])
```

```python
def recommend(name,n):
    sim_distance=linear_kernel(vectors[index[name]],vectors)
    sim_score_df=pd.DataFrame(sim_distance).T
    sim_score_df.columns=['Scores']
    sim_score_df=sim_score_df.sort_values("Scores", ascending=False)

    list=[]

    for i in range(0,n+1):
        result={'Title': movies_data["Title"][sim_score_df.index[i]],
                'Genre': movies_data["Genre"][sim_score_df.index[i]],
                'Rating': movies_data["Rating"][sim_score_df.index[i]]


        }

        list.append(result)

    return list
```

```python
Recommendation_df = pd.DataFrame(recommend('Dangal',5))
Recommendation_df
```
